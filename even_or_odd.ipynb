# Project: Check Even or Odd

```python
# This project helps beginners learn how to use if-else statements and the modulus operator.
```

```python
# Step 1: Get input from the user
num = input("Enter an integer number: ")
```

```python
# Step 2: Convert input to integer
num = int(num)
```

```python
# Step 3: Check if the number is even or odd
if num % 2 == 0:
    print(f"{num} is an even number.")
else:
    print(f"{num} is an odd number.")
```
